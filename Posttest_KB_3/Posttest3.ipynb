***IMPORT LIBRARY***

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

***MEMASUKKAN DATASET***

```python
df = pd.read_csv("players_stats_by_season_full_details.csv")
```

***PIE CHART***

```python
Stages = df.dropna(subset=['Stage']).sort_values(by=['Stage'])['Stage'].unique()

total = df.groupby(['Stage'])['GP'].count()

patches, texts, pcts = plt.pie(total, labels=Stages, autopct='%.1f%%', wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       textprops={'size': 'x-large'})


plt.title('Total Pemain Basker Berdasarkan Stage')
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```

Pada diagram pie chart ini menghitung dan memvisualisasikan distribusi jumlah pemain basket berdasarkan tahap (stage)


***HEAT MAP***

```python
columns_to_plot = ['3PM', '3PA', 'MIN', 'FGM', 'FGA']
correlation_matrix = df[columns_to_plot].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi antara 3PM, 3PA, MIN, FGM, dan FGA')
plt.show()
```

heatmap yang menunjukkan seberapa kuat korelasi antara kolom-kolom yang disebutkan. Setiap sel di heatmap berisi angka yang menggambarkan korelasi antara dua kolom. Angka-angka ini berkisar antara -1 (korelasi negatif sempurna) hingga 1 (korelasi positif sempurna), dengan 0 menunjukkan tidak adanya korelasi. 


***BAR CHART***

```python
league_counts = df['League'].value_counts()
plt.figure(figsize=(12, 6))
league_counts.plot(kind='bar', color='skyblue')
plt.title('Jumlah Pemain dalam Setiap Liga')
plt.xlabel('Liga')
plt.ylabel('Jumlah Pemain')
plt.show()
```

bar chart yang memvisualisasikan jumlah pemain dalam setiap liga dalam dataset. Setiap batang pada chart mewakili satu liga, dan tinggi batang tersebut menunjukkan jumlah pemain dalam liga tersebut.

***HISTOGRAM***

```python
gp_df = df['GP']
plt.figure(figsize=(10, 6))
plt.hist(gp_df, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Jumlah Games Played (GP)')
plt.ylabel('Jumlah Pemain')
plt.title('Histogram Jumlah Games Played (GP) Pemain Basket')
plt.tight_layout()
plt.show()
```

histogram yang menggambarkan distribusi jumlah permainan yang dimainkan oleh para pemain basket dalam dataset. Histogram ini membantu Anda untuk memahami sebaran data, seperti berapa banyak pemain yang memainkan sejumlah permainan tertentu dalam dataset.
