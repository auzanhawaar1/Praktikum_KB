```python
record, attribute = basket.shape
jml_object = basket.select_dtypes(["float", "int"]).any().sum()
nama_attribute = basket.columns.values
jml_null = basket.isnull().any().sum()

info = {"Jumlah Record": record, 
        "Jumlah Attribute" : attribute, 
        "Nama Attribute" : list(nama_attribute), 
        "Jumlah Object" : jml_object, 
        "Jumlah Nilai Null": jml_null}

for x in info :
    print(x, ":", info[x])
```
